```python
import numpy as np
from pylab import *
import scipy
import pandas as pd
import datetime
from scipy import optimize

growth_curves = pd.read_excel (r'ecAG_16_17_Dex_30C_PL1.xlsx', sheet_name='Raw data')

start=datetime.time(0,0,0)
measure_times=[(datetime.datetime.combine(datetime.date.today(), growth_curves['Time'][i])- datetime.datetime.combine(datetime.date.today(), start)).total_seconds() / 3600  for i in range(len(growth_curves))]
```

```python
ecAG16_0_columns=['D1','D3','D5','D7','D9','D11']
ecAG16_15_columns=['B1','B3','B5','B7','B9','B11']
ecAG16_30_columns=['G1','G3','G5','G7','G9','G11']
ecAG16_45_columns=['F1','F3','F5','F7','F9','F11']
ecAG16_60_columns=['C1','C3','C5','C7','C9','C11']
ecAG16_75_columns=['E1','E3','E5','E7','E9','E11']
ecAG16_90_columns=['A1','A3','A5','A7','A9','A11']
ecAG16_105_columns=['H1','H3','H5','H7','H9','H11']

ecAG17_0_columns=['D2','D4','D6','D8','D10','D12']
ecAG17_15_columns=['B2','B4','B6','B8','B10','B12']
ecAG17_30_columns=['G2','G4','G6','G8','G10','G12']
ecAG17_45_columns=['F2','F4','F6','F8','F10','F12']
ecAG17_60_columns=['C2','C4','C6','C8','C10','C12']
ecAG17_75_columns=['E2','E4','E6','E8','E10','E12']
ecAG17_90_columns=['A2','A4','A6','A8','A10','A12']
ecAG17_105_columns=['H2','H4','H6','H8','H10','H12']

curves_ecAG16_0=np.log(growth_curves.loc[:, ecAG16_0_columns])
curves_ecAG16_15=np.log(growth_curves.loc[:, ecAG16_15_columns])
curves_ecAG16_30=np.log(growth_curves.loc[:, ecAG16_30_columns])
curves_ecAG16_45=np.log(growth_curves.loc[:, ecAG16_45_columns])
curves_ecAG16_60=np.log(growth_curves.loc[:, ecAG16_60_columns])
curves_ecAG16_75=np.log(growth_curves.loc[:, ecAG16_75_columns])
curves_ecAG16_90=np.log(growth_curves.loc[:, ecAG16_90_columns])
curves_ecAG16_105=np.log(growth_curves.loc[:, ecAG16_105_columns])

curves_ecAG17_0=np.log(growth_curves.loc[:, ecAG17_0_columns])
curves_ecAG17_15=np.log(growth_curves.loc[:, ecAG17_15_columns])
curves_ecAG17_30=np.log(growth_curves.loc[:, ecAG17_30_columns])
curves_ecAG17_45=np.log(growth_curves.loc[:, ecAG17_45_columns])
curves_ecAG17_60=np.log(growth_curves.loc[:, ecAG17_60_columns])
curves_ecAG17_75=np.log(growth_curves.loc[:, ecAG17_75_columns])
curves_ecAG17_90=np.log(growth_curves.loc[:, ecAG17_90_columns])
curves_ecAG17_105=np.log(growth_curves.loc[:, ecAG17_105_columns])

IPTG_concs=[0,15,30,45,60,75,90,105]
curves_ecAG16=[curves_ecAG16_0,curves_ecAG16_15,curves_ecAG16_30,curves_ecAG16_45,curves_ecAG16_60,curves_ecAG16_75,curves_ecAG16_90,curves_ecAG16_105]
curves_ecAG17=[curves_ecAG17_0,curves_ecAG17_15,curves_ecAG17_30,curves_ecAG17_45,curves_ecAG17_60,curves_ecAG17_75,curves_ecAG17_90,curves_ecAG17_105]
```

```python
fig = plt.figure(figsize=(25,15))
fig.subplots_adjust(hspace=0.4, wspace=0.25)
fig.suptitle('ecAG16',fontsize=22)
for i in range(len(IPTG_concs)):
    subplot = fig.add_subplot(2, 4, i+1)
    subplot.plot(measure_times,curves_ecAG16[i])
    subplot.set_xlabel('Time (hours)')
    subplot.set_ylabel('OD600')
    subplot.set_title('IPTG '+str(IPTG_concs[i])+' uM')
plt.show()
```

```python
fig = plt.figure(figsize=(25,15))
fig.subplots_adjust(hspace=0.4, wspace=0.25)
fig.suptitle('ecAG17',fontsize=22)
for i in range(len(IPTG_concs)):
    subplot = fig.add_subplot(2, 4, i+1)
    subplot.plot(measure_times,curves_ecAG17[i])
    subplot.set_xlabel('Time (hours)')
    subplot.set_ylabel('OD600')
    subplot.set_title('IPTG '+str(IPTG_concs[i])+' uM')
plt.show()
```

```python
range_inf=10
range_sup=18

I=0
while 1:
    if measure_times[I]<=range_inf:
        I=I+1
    else:
        break
        
J=0
while 1:
    if measure_times[J]<=range_sup:
        J=J+1
    else:
        break
```

```python
ecAG16_growth_0=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_0[i][I:J].tolist())[0][1] for i in ecAG16_0_columns]

ecAG16_growth_15=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_15[i][I:J].tolist())[0][1] for i in ecAG16_15_columns]

ecAG16_growth_30=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_30[i][I:J].tolist())[0][1] for i in ecAG16_30_columns]

ecAG16_growth_45=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_45[i][I:J].tolist())[0][1] for i in ecAG16_45_columns]

ecAG16_growth_60=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_60[i][I:J].tolist())[0][1] for i in ecAG16_60_columns]

ecAG16_growth_75=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_75[i][I:J].tolist())[0][1] for i in ecAG16_75_columns]

ecAG16_growth_90=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_90[i][I:J].tolist())[0][1] for i in ecAG16_90_columns]

ecAG16_growth_105=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG16_105[i][I:J].tolist())[0][1] for i in ecAG16_105_columns]

ecAG16_growth_rates=[ecAG16_growth_0,ecAG16_growth_15,ecAG16_growth_30,ecAG16_growth_45,ecAG16_growth_60,ecAG16_growth_75,ecAG16_growth_90,ecAG16_growth_105]


ecAG17_growth_0=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_0[i][I:J].tolist())[0][1] for i in ecAG17_0_columns]

ecAG17_growth_15=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_15[i][I:J].tolist())[0][1] for i in ecAG17_15_columns]

ecAG17_growth_30=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_30[i][I:J].tolist())[0][1] for i in ecAG17_30_columns]

ecAG17_growth_45=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_45[i][I:J].tolist())[0][1] for i in ecAG17_45_columns]

ecAG17_growth_60=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_60[i][I:J].tolist())[0][1] for i in ecAG17_60_columns]

ecAG17_growth_75=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_75[i][I:J].tolist())[0][1] for i in ecAG17_75_columns]

ecAG17_growth_90=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_90[i][I:J].tolist())[0][1] for i in ecAG17_90_columns]

ecAG17_growth_105=[scipy.optimize.curve_fit(lambda t,a,b: a+b*t, measure_times[I:J], curves_ecAG17_105[i][I:J].tolist())[0][1] for i in ecAG17_105_columns]

ecAG17_growth_rates=[ecAG17_growth_0,ecAG17_growth_15,ecAG17_growth_30,ecAG17_growth_45,ecAG17_growth_60,ecAG17_growth_75,ecAG17_growth_90,ecAG17_growth_105]
```

```python
plt.boxplot(ecAG16_growth_rates)
plt.xticks([1,2,3,4,5,6,7,8],IPTG_concs)
plt.title('ecAG16')
plt.xlabel('IPTG concentration (uM)')
plt.ylabel('Growth rate (1/h)')
plt.show()
```

```python
plt.errorbar(IPTG_concs,map(mean,ecAG16_growth_rates),map(std,ecAG16_growth_rates))
plt.title('ecAG16')
plt.xlabel('IPTG concentration (uM)')
plt.ylabel('Growth rate (1/h)')
plt.show()
```
